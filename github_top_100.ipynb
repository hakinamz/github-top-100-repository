```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```
df = pd.read_csv('Most starred Github Repositories.csv')
```

```
df.info()
```

```
df.convert_dtypes().head()
```

```
df['Language'].unique()
```

```


sns.set_style('darkgrid')

fig, ax = plt.subplots(figsize = (15,10))

df['Language'].value_counts().plot(kind = 'bar', color = 'purple', alpha = 0.7)
plt.xticks(rotation = 45)
plt.ylabel('Number of repositories')
plt.xlabel('Language')
plt.title('Most used Languages in the Github Top 100', fontsize = 18)

plt.show()
```

```
X = df['Language']
y = df['Stars']


fig, ax = plt.subplots(figsize = (15,10))

plt.scatter(X, y, s = 90, alpha = 0.7)
plt.xticks(rotation = 45)
plt.ylabel('Number of Stars')
plt.xlabel('Language')
plt.title('Number of Stars by Repositorie', fontsize = 18)

plt.show()
```

```
y = df['Forks']


fig, ax = plt.subplots(figsize = (15,10))

plt.scatter(X, y, s = 90, alpha = 0.7)
plt.xticks(rotation = 45)
plt.ylabel('Number of Forks')
plt.xlabel('Language')
plt.title('Number of Forks by Repositorie', fontsize = 18)

plt.show()
```
